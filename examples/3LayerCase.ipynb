```python
import numpy as np
from saem import CSEMData
```

```python
f = [10, 100, 1000]
x = np.arange(10., 3001, 10)
```

```python
txLen = 1000  # length of the transmitter
altitude = 100
self = CSEMData(f=f, rx=x, txPos=np.array([[0, 0], [-txLen/2, txLen/2]]), alt=altitude)
print(self)
self.showPos();
```

```python
rho = [1000, 1000, 1000]
thk = [100, 100]
self.cmp = [1, 0, 1]
self.simulate(rho=rho, thk=thk)
```

```python
self.showLineData(what="response");
```

```python
self.showLineFreq(line=1, nf=1, what="response")
```
